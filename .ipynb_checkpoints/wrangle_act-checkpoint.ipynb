#### Code

```python
#this line of code creates a more descriptive name for the rating_numerator column
df_tae_clean.rename(columns = {'rating_numerator': 'rating_over_10'}, inplace = True)
```

```python
#this line of code converts the rating column to float datatype
df_tae_clean.rating_over_10 = df_tae_clean.rating_over_10.astype(float)
```

*Next, there appears to be some ratings that are floating point numbers hence , i will be writing a code to extract this numbers and uppdate my rating columns*

```python
#this line of code checks for tweet id with a floating point number as rating
df_tae_clean[df_tae_clean.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_over_10']]
```

*The above code shows index of where we have float type numbers as our rating.Next, we then try to correctly extract the rating correctly below*

```python
#extract the correct rating from the text
values['new'] = df_tae_clean.loc[df_tae_clean.text.str.contains('(\d+\.\d+\/)'), 'text'].str.extract('(\d+\.\d+)')

values
```

```python
values['new']
```

```python
#this line f code updates the dataframe with the newly extracted rating numerator
df_tae_clean.loc[df_tae_clean.text.str.contains('(\d+\.\d+\/)'), 'rating_over_10'] = values['new']
```

```python
#this line of code displays characteristics of the remaining outliers
df_tae_clean[df_tae_clean.rating_over_10 > 15]
```

*It happens that some values greater than 15 are not for one dog instead, they are for a squad or group of dogs. I will also be resolving this issue by trying to correctly get the normalrating for one dog*

```python
# this for loop finds any text that has two ratings per tweet_id.
for index in df_tae_clean.tweet_id:
    # Extracting the text of each tweet
    txt = df_tae_clean[df_tae_clean.tweet_id == index].text.tolist()[0]
    # This regex will find the pattern YY/YY.
    rating = re.findall('\d+\d\/\d+', txt)
    # If any rating is a list of two elements.
    if len(rating) == 2:
        if(rating[0] == '50/50'):
            #I want only the first part of the second element.
            numerator = rating[1].split('/')[0]
            denominator = rating[1].split('/')[1]
            #I use the .loc() to update the data alocated in df_tae_clean
            df_tae_clean.loc[df_tae_clean[df_tae_clean['tweet_id'] == index].index,'rating_over_10'] = float(numerator)
```

*Upon visually assessing some of this outliers, i will be dropping three twee ids whose rating are either unussually large or the rating extracted was actually not a rating*

```python
#this line of code resets the index of the dataframe
df_tae_clean = df_tae_clean.reset_index(drop = True)

#next, i want to drop the identified outlier ratings that are not ratings or are just too large using their row index
df_tae_clean.drop([403, 770, 1821], inplace = True)
```

*Next, i will be writing a code to properly return the rating for each dog rating where the dogs were grouped together*

```python
outliers = df_tae_clean[df_tae_clean.rating_over_10 > 15]#i assigned the outliers to a variable

for index in outliers.tweet_id:#used a for loop to use the twitter id for every outlier to recalculate the rating for each dog
    
    new_values = (outliers.rating_over_10/outliers.rating_denominator)*10
    #i then assigned the newly calculated values to replace the old one
    df_tae_clean.loc[df_tae_clean[df_tae_clean['tweet_id'] == index].index,'rating_over_10'] = new_values
```

## Comparing the performance of algorithm's predictions(p1,p2 and p3)

*Given that the image of each dog was ran through a neural network to predict if the image is a dog or not and also to return the confidence of each prediction and its percieved breed*

*here, i will like to analyse and compare the thre predictions and how they differ. Firstly, i will define a function to plot dog breeds where the number of dog breeds predicted has more than 20 occurences for each predition ,that is where the prediction is truly a dog(p1_dog = TRUE)*

```python
#define the required function
def dog_pred(df_r, title, xlabel, ylabel):#df_r is dataframe passed into  the function, title is the proposed 
                                                            #title of the chart, font and font_title specifies the size of the fonts
    
    df_r[df_r > 20].plot(kind = 'barh', figsize = (10,5), color = "#C13016", edgecolor = ['Black']*len(df_r[df_r > 20].value_counts()));#plots a bar chart 
    
    plt.title(label = title, fontsize = 16);#prints the title 
    
    plt.xlabel(xlabel = xlabel, fontsize = 12);#assigns a xlabel
    
    plt.ylabel(ylabel = ylabel, fontsize = 12);#assigns a ylabel
    
    
```

**Prediction Number 1(p1)**

```python
#returns the number of dog breeds found in the first prediction
df_master[df_master.p1_dog].p1.value_counts()
```

```python
#this line of code returns the number of occurences for each dog breed where p1_dog = true
df_p1 = df_master[df_master.p1_dog].p1.value_counts()

#calling the function to plot the dog breeds that have more than 15 occurences
dog_pred(df_p1,
        title = "Dog breeds with more than 20 occurences for each prediction(p1)",
         xlabel = "Number of Occurences",
         ylabel = "Dog Breeds"
        )
```

#### Prediction Number 2(p2)

```python
#returns the number of dog breeds found in the second prediction
df_master[df_master.p2_dog].p2.value_counts()
```

```python
#this line of code returns the number of occurences for each dog breed where p2_dog = true
df_p2 = df_master[df_master.p2_dog].p2.value_counts()

#calling the function to plot the dog breeds that have more than 15 occurences
dog_pred(df_p2,
        title = "Dog breeds with more than 20 occurences for each prediction(p2)",
         xlabel = "Number of Occurences",
         ylabel = "Dog Breeds"
        )
```

### Prediction Number 3(p3)

```python
#returns the number of dog breeds in the 3rd prediction
df_master[df_master.p3_dog].p3.value_counts()
```

```python
#this line of code returns the number of occurences for each dog breed where p3_dog = true
df_p3 = df_master[df_master.p3_dog].p3.value_counts()

#calling the function to plot the dog breeds that have more than 15 occurences
dog_pred(df_p3,
        title = "Dog breeds with more than 20 occurences for each prediction(p3)",
         xlabel = "Number of Occurences",
         ylabel = "Dog Breeds"
        )
```

*Now, the next set of operations aims to find the number of dog breeds in each prediction where the dog breed have more than 20 occurences*

```python
# Calculating the number of breeds in each algorithm with more than 20 occurencies.
p1_qty = sum(df_master[df_master.p1_dog].p1.value_counts() > 20)
p2_qty = sum(df_master[df_master.p2_dog].p2.value_counts() > 20)
p3_qty = sum(df_master[df_master.p3_dog].p3.value_counts() > 20)

# Printing a summary of occurencies.
print("P1: {} breeds.\nP2: {} breeds.\nP3: {} breeds.".format(p1_qty,p2_qty,p3_qty))
```

*The operation below aims to find the total number of dog in the breeds mentioned above*

```python
# Calculating the number of dogs classified using the threshold of 20.
dog_qty_20_p1 = df_master[df_master.p1_dog].p1.value_counts()[df_master[df_master.p1_dog].p1.value_counts() > 20].sum()
dog_qty_20_p2 = df_master[df_master.p2_dog].p2.value_counts()[df_master[df_master.p2_dog].p2.value_counts() > 20].sum()
dog_qty_20_p3 = df_master[df_master.p3_dog].p3.value_counts()[df_master[df_master.p3_dog].p3.value_counts() > 20].sum()

# Printing a summary of occurencies.
print("P1: {} dogs.\nP2: {} dogs.\nP3: {} dogs.".format(dog_qty_20_p1,dog_qty_20_p2,dog_qty_20_p3))
```

*From the analysis above, i can say that the number one prediction appears to keep it concentration on a few breeds while having the highest number of dogs.This compared to the 3rd prediction which has a lot of breeds with more than 20 occurences and the lowest number of dogs*
